```python
movies.columns
```

```python
movies=movies[['id','title','genre','overview']]
movies
```

```python
movies['tags']=movies['genre']+movies['overview']
movies
```

```python
new=movies.drop(columns=['genre','overview'])
new
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words='english')
cv
```

```python
#vector=cv.fit_transform(new['tags'].values.astype('U')).toarray()
vector = cv.fit_transform(new['tags'].values.astype('U')).toa1rray()
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity
#similarity.shape
```

```python
new[new['title']=="The Godfather"].index[0]
#here we want to get the movie names by index vaue so we checked the index value of certain movie eith the above instructio
```

```python
#enumerate us vector ko iterate karega uski index value or values ke saath
list(enumerate(similarity[2]))
```

```python
#doubt key=lambda vector:vector[1]
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
  print(new.iloc[i[0]].title)
```

```python
def recommend(movies):
  index=new[new['title']==movies].index[0]
  distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
  for i in distance[0:5]:
    print(new.iloc[i[0]].title)
```

```python
recommend("Dilwale Dulhania Le Jayenge")
```
